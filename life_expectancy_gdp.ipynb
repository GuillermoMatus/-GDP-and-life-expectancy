```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels
import statsmodels.api as sm
import math
```

```python
life_expectancy_gdp = pd.read_csv("F://Luis//Codecademy//Life-Expectancy-and-GDP-Starter//Life-Expectancy-and-GDP-Starter//all_data.csv")
column_names = list(life_expectancy_gdp.columns.values)
print(column_names)
print(life_expectancy_gdp.head())
gdp = life_expectancy_gdp
```

```python
countries = list(set(gdp["Country"]))
#print(countries)
years = list(set(gdp["Year"]))
#print(years)
number_of_data = len(gdp)
years_resume = gdp["Life expectancy at birth (years)"].describe()
print(years_resume)

gdp_resume = gdp["GDP"].describe()
print(gdp_resume)
```

```python
print("The number of data in this table is: ", number_of_data)
print("The years are from ", gdp["Year"].min()," to ", gdp["Year"].max())
print("this is the list of countries that participate:")
for country in countries:
    print (country)
```

```python
gdp = gdp.rename(columns= {"Life expectancy at birth (years)": "LEAB"})

for c in countries:
    gdpc = gdp.loc[:,"Country"] == c
    gdpa = gdp.loc[gdpc]
    sns.lmplot(x="Year",y="LEAB",data = gdpa)
    plt.title(c)
    plt.show()
    plt.clf()
```

```python
sns.lmplot(x="Year",y="LEAB",data = gdp, hue = "Country")
plt.show()
plt.clf()
sns.lmplot(x="Year",y="GDP",data = gdp, hue = "Country")
plt.show()
plt.clf()
sns.lmplot(x="GDP",y="LEAB",data = gdp, hue = "Country")
plt.show()
plt.clf()
```
